## **2. MNIST 데이터 불러오기**

## **3. 신경망 구축**

```python
from tensorflow.keras import models
from tensorflow.keras import layers

network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))
```

1. 손실함수 (loss function) : 네트워크가 train 데이터에서 얼마나 학습을 잘 수행하는지 측정, 올바른 방향으로 조정할 수 있다.
2. 옵티마이저 (optimizer) : 최적화 알고리즘
3. test와 train을 모니터링 할 Metrics. 여기서는 정확성만을 신경쓴다.

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

1. 옵티마이저를 rmsprop로 설정.
2. 손실함수를 categorical_crossentropy로 설정.

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

1. 3차원 데이터를 2차원 데이터로 차원수를 재설정한다. | reshape() 차원의 크기 변경
2. 0 ~ 1 값으로 스케일링 | astype() 데이터 타입 변경

```python
from tensorflow.keras.utils import to_categorical

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

1. 범주형 데이터로 변환한다.

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

1. train 데이터에 적용한다.
2. train 데이터에서 정확도 98.8%로 나타난다.

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
```

```python
print('test_acc:', test_acc)
```

1. test 데이터에 적용한다.
2. test 데이터에서도 정확도 97.8%로 매우 높게 나타난다.
